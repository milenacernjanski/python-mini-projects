# Break
    -- a break causes the interpreter to break out of the while loop

```python
subtotal = 0     # running subtotal of inputs
user_input = 1   # initial input to make while loop work
input_num = 0    # number of inputs by user

while user_input >= 0:
  user_input = float(input('Please enter a number:'))

  if user_input < 0:
    break            # break out of while loop

  input_num += 1   # number of inputs by user
  print('User number:', user_input)

  subtotal += user_input
  print('Subtotal:', subtotal)

  if input_num >= 0:        # == no =
    print('Average', subtotal / (input_num))

print('Good-bye')
```

# Continue
    -- once the interpreter sees continue, it jumps to the top of the while loop, ignoring the rest of the while-loop code below

```python
subtotal = 0     # running subtotal of inputs
user_input = 1   # initial input to make while loop work
input_num = 0    # number of inputs by user

while user_input >= 0:
  user_input = float(input('Please enter a number:'))

  if user_input < 0:
    break            # break out of while loop

  input_num += 1   # number of inputs by user
  print('User number:', user_input)

  subtotal += user_input
  print('Subtotal:', subtotal)

  if user_input == 999:
    continue     # go immediately to top of while loop

  if input_num >= 0:        # == no =
    print('Average', subtotal / (input_num))

print('Good-bye')


# x += 2  x = x + 2
# x -= 4  x = x - 4
# x *= 2 x = x * 2
# there must be x = 'something' otherwise it says it is undefined and program cannot work
```

# for loop

```python
# Initialize variables
total_cost = 0.0
num_tickets = 4

# Loop through each family member
for i in range(num_tickets):
    age = int(input(f"Enter the age of family member {i + 1}: "))

    # Check the age and calculate the ticket cost
    if age < 3:
        ticket_cost = 0.00
    elif 3 <= age <= 12:
        ticket_cost = 10.25
    else:
        ticket_cost = 15.50

    # Update the total cost
    total_cost += ticket_cost

    # Display the running total
    print(f"Your cost before taxes is currently ${total_cost:.2f}")

# Calculate the final total cost including sales tax
tax_rate = 0.075  # 7.5% sales tax
final_cost = total_cost * (1 + tax_rate)

# Display the final total cost
print(f"Final total cost including 7.5% sales tax: ${final_cost:.2f}")
```

```python
print('Movie Theater of Milena Cernjanski')
print('Cost of Movie Tickets For Family of Four')

# Initialize variables
running_total = 0.0
num_tickets = 4

# Loop through each family member
for i in range(num_tickets):
    age = int(input(f"Enter the age of family member {i + 1}: "))

    # Check the age and calculate the ticket cost
    if age < 3:
        ticket_cost = 0.00
    elif 3 <= age <= 12:
        ticket_cost = 10.25
    else:
        ticket_cost = 15.50

    # Update the total cost
    running_total += ticket_cost

    # Display the running total
    print(f"Your cost before taxes is currently ${running_total:.2f}")

# Calculate the final total cost including sales tax
tax_rate = 0.075  # 7.5% sales tax
final_cost = running_total * (1 + tax_rate)

# Display the final total cost
print(f"Final total cost including 7.5% sales tax: ${final_cost:.2f}")

#*********************************************************************
# Extra Credit
#*********************************************************************

# Initialize variables
total_cost = 0.0
num_tickets = 0

# Loop to input ages and calculate ticket costs
while True:
    age_input = input(f"Enter the age of family member {num_tickets + 1} (or 'done' to calculate total): ")

    if age_input.lower() == 'done':
        break

    try:
        age = int(age_input)
    except ValueError:
        print("Invalid input. Please enter a valid age or 'done' to calculate the total.")
        continue

    if age < 3:
        ticket_cost = 0.00
    elif 3 <= age <= 12:
        ticket_cost = 10.25
    else:
        ticket_cost = 15.50

    total_cost += ticket_cost
    num_tickets += 1
    print(f"Your cost before taxes is currently ${total_cost:.2f}")

# Calculate the final total cost including sales tax
tax_rate = 0.075  # 7.5% sales tax
final_cost = total_cost * (1 + tax_rate)

# Display the final total cost
print(f"Total cost for {num_tickets} tickets including 7.5% sales tax: ${final_cost:.2f}")
```
