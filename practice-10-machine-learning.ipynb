## Data Cleaning

```python
###################################################################################
# Revert to a clean training set and separate out the target column of data
# -- We want to do transformations only on the predictors, not thre traget data
###################################################################################

# housing will now equal the first strat_train_set without median_house data

housing = strat_train_set.drop("median_house_value", axis=1) # 1 for column

# housing_labels now just the target median_house_value data

house_labels = strat_train_set["median_house_value"].copy()
```

## Three Options for cleaning a column containing NaN data (Null)

    Option 1: Get rid of rows with NaN data
    Option 2: Get rid of the column with NaN data
    Option 3: Change the NaN to some value (median, mean, constant
        -- Less flexible solution( for option 3)
        -- Better solution (for option 3)


```python
###################################################
# Identify the rows with NaN data

null_rows_idx = housing.isnull().any(axis=1)
housing.loc[null_rows_idx].head()
```

```python
####################################################################################
# Option 1 (Gets rid of some rows): Get rid of rows with no total_bedrooms data
#              .dropna()
####################################################################################

housing_option1 = housing.copy()
housing_option1.dropna(subset=["total_bedrooms"], inplace=True)
housing_option1.loc[null_rows_idx].head()
```

```python
####################################################################################
# Option 2 (Gets rid of a column of data)
#       .drop()
####################################################################################

housing_option2 = housing.copy()
housing_option2.drop("total_bedrooms", axis=1, inplace=True)
housing_option2.loc[null_rows_idx].head()
```

```python
####################################################################################
# Option 3 (Better: Change missing data to value) --> but not the best version
####################################################################################

housing_option3 = housing.copy()

median = housing["total_bedrooms"].median()
housing_option3["total_bedrooms"].fillna(median, inplace=True)

housing_option3.loc[null_rows_idx].head()
```

```python
####################################################################################
####################################################################################
# BEST WAY to do Option 3: SimpleImputer
#   -- "imputation" = change null value to another value like median
#
#            For numeric data only
#               -- strategy="median"
#               -- strategy="mean"
#
#            For numeric and non-numeric data (like categorical data)
#               -- strategy="most_frequent"
#               -- strategy="constant", fill_value=...
#   -- imputer is an instance of a SimpleImputer class to turn nulls to median values
####################################################################################
####################################################################################

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median") # create an instance: imputer
imputer.strategy
```

```python
####################################################################################
# Because using "median" strategy, create a copy of data with only numeric attributes.
####################################################################################

housing_num = housing.select_dtypes(include=[np.number])
```

```python
####################################################################################
# Fit the imputer instance to the training data
#     -- imputer computes the median value of each attribute and stores its results in statistics_
####################################################################################

imputer.fit(housing_num)
```

```python
############################
# Check manually
############################

housing_num.median().values
```

```python
imputer.statistics_
```

```python
housing_num.head()
```

```python
#######################################################################################
# Transform the training set by replacing missing values with the learned median values
#
# NOTE: The output is NumPy array without column names nor index
#######################################################################################

X = imputer.transform(housing_num)
```

```python
imputer.feature_names_in_
```

```python
#######################################################################################
# Wrap X in a DataFrame to recover the column names and index from housing_num
#######################################################################################

housing_tr = pd.DataFrame(X, columns=housing_num.columns,
                          index=housing_num.index)
```

```python
###################################################
# Check to see if null values are changed
###################################################

housing_tr.loc[null_rows_idx].head()
```

## Feature Scaling

```python
#########################################################################
# Feature Scaling
#   -- Two common ways to do Feature Scaling:
#         -- min-mix (normalization)
#             -- For neural networks, zero-mean inputs, prefer -1 to 1
#         -- standardization
#             -- First subtracts the mean value
#                   -- (so standardizes values to a zero mean)
#             -- Then, divides the result by the standard deviation
#                   -- (so standardized values have a standard deviation equal to 1)
#
#   -- Standardization does not restrict values like min-max does
#   -- Standardization is much less affected by outliers.
#########################################################################

###########################
# min-max (normalization)
###########################

from sklearn.preprocessing import MinMaxScaler

#############################################
# -1 and 1 for neural networks if we want

min_max_scaler = MinMaxScaler(feature_range=(-1,1))
housing_num_min_max_scaled = min_max_scaler.fit_transform(housing_num)
```
