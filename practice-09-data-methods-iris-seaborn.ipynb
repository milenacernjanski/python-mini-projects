# Inheritance
    Parent / Child relationship
    

```python
class Employee:   # Parent Class

  location = "Found, Mississippi"  # class variable

  # Constructor - Initializing object with instantiation variables
  def __init__(self, name, phone, address, email, role, payrate):
    self.name = name
    self.phone = phone
    self.address = address
    self.email = email
    self.role = role
    self.payrate = payrate

  def get_info(self):
    return (f'{self.name}, {self.phone}, {self.address}, {self.email}, {self.role}, {self.payrate}')

class Developer(Employee):

  # Constructor

  def __init__(self, name, phone, address, email, role, payrate, program_language):
    super().__init__(name, phone, address, email, role, payrate)  # Get Parent stuff
    self.program_language = program_language

  def get_info(self):  # method overriding of Parent class method
    return (f'{self.name}, {self.payrate}, {self.program_language}')

employee1 = Developer("Sally", "455-234-8989", "Lost and Found, Tennessee", "s@BigStore.com", "Account Designer", 999, "Python")


print(employee1.get_info())

employee2 = Employee("It", "000-000-0000", "Dairy, Maine", "i@BigStore.com", "Clown", 2 )


print(employee2.get_info())
```

```python
class Players:

  def __init__(self, name, height, weight, jerseynum, position, payrate):
    self.name = name
    self.height = height
    self.weight = weight
    self.jerseynum = jerseynum
    self.position = position
    self.payrate = payrate

  def get_info(self):
    return(f'{self.name}, {self.height}, {self.weight}, {self.jerseynum}, {self.position}, {self.payrate}')

class Developer(Employee):

  def __init__(self, name, height, weight, jerseynum, position, payrate, best_award):
    super().__init__(name, height, weight, jerseynum, position, payrate)
    self.best_award = best_award

  def get_info(self):
    return (f'{self.name}, {self.payrate}, {self.best_award}')

player1 = Developer("Milena Cernjanski", 1.73, 60, "#5", "libero", "50,000.00", "MVP")

print(player1.get_info())
```

# Useful built-in Functions and Methods

## Modifiying a value Function

```python
x = 3.1459
y = -3.1459
z = 1234.56789
q = 0.12345

# abs(x)   -- Absolute value
print(abs(x))
print(abs(y))

# round(x)   -- Round to nearest integer
print(round(x))
print(round(z))

# round(x,2)  -- Rouund to 2 digits after the decimal
print(round(x,2))
print(round(q,3))
```

## List Functions



```python
x_list = [4, 5, 9, 8, 2]
y_list = ['watremelon', 'taco', 'popcorn']

# len(x_list)     # mength of list
print(len(x_list))

# min(x_list)     # minimum value
print(min(x_list))

# max(x_list)     # maximum value in a list
print(max(x_list))

# sorted(x_list)  # sort list!!!!!!!!!!
print(sorted(x_list))
print(sorted(y_list))   # by alphabet

# sorted(x_list, reverse=True)
print(sorted(x_list, reverse=True))
print(sorted(y_list, reverse=True))


# sorted(y_list, key=len)
print(sorted(y_list, key=len))

print(sorted(y_list, key=len, reverse=True))
```

## Object-Oriented Stle of Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Years':['2000','2002','2004','2006','2008','2010', '2012','2014','2016'],
        'Men':[189.1, 191.8, 193.6, 196.3, 194.4, 197.0, 197.8, 199.2, 170.4],
        'Women':[175.7, 176.4, 176.5, 176.2, 175.9, 175.9, 175.7, 175.8, 175.3]}

heights_df = pd.DataFrame(data)

# fig: object for the whole plot
    # figs is an istance of the Figure class
# ax: object that holds subplots within the figure
    # ax is an instance of the AxesSubplot class
fig, ax =plt.subplots()

# method on the ax object to create the subplot "Men"
# .plot(x,y, label='...' )
ax.plot(heights_df['Years'], heights_df['Men'], label = 'Men')

# method on the ax object to create the subplot "Women"
ax.plot(heights_df['Years'], heights_df['Women'], label = 'Women')

ax.set_xlabel('Years')
ax.set_ylabel('Height (cm)')
ax.set_title('Average heights of Men and Women over time')
ax.legend()   # method on the ax object that creates legend

plt.show()  # display the whole graph --> IMPORTANT IF NOT IN JUPYTER NOTEBOOK
```

## Object-Oriented with Multiple Plots

```python
fig, (ax1, ax2) = plt.subplots(1,2) # one figure with two axes subplots

ax1.plot('Years','Women', data=heights_df)
ax1.set_xlabel('Year')
ax1.set_ylabel('Height (cm)')
ax1.set_title('Women')
ax1.legend(['Women'])

ax2.plot('Years','Men', data=heights_df)
ax2.set_xlabel('Year')
ax2.set_title('Men')
ax2.legend(['Men'])

fig.autofmt_xdate(rotation=65)
plt.show()
```
