# if

```python
x = 5
if x < 7:    # must have a colon :
  print(x)    # must be indented
  print(x+1)

print('done')

if x < 3:
  print(x)    # when it is not true, it ignores all the indented*****
  print(x+1)
                                # space does not matter, only indention
  print('done')

if x > 3:
  print('x is greater than 3')
  print(x)
  print(x-2)

if x == 5:       # double equal sign is =
  print('x is equal to 5')

if x != 4:      # != means does not equal
  print('x does not equal 4')

if x <=8:
  print('x is less than or equal to 8')
```

# if else

```python
x = 5
if x != 5:
  print('x != 5')
else:
  print('x == 5')

if x >= 4:
  print('x >= 4')
else:
  print('x is not >= 4')
```

# if - elif - else


```python
x = 9    # one = means "assigns value from rigth side"
if x == 8:
  print('x == 8')   # two == means "equivalent" or "equals"
elif x < 8:
  print ('x < 8')
else:
  print('x > 8')

# == equals
# != not equals

x = 5
if x == 15:
  print ('x == 15')
elif x > 15:
  print('x < 15')
else:
  print('x != 15')

```

# while loop

```python
x = 1

#while x < 5:    # infinite loop
  #print(x)

while x <= 5:     # also do x < 6
  print(x)
  x = x + 15

x = 10
while x <= 20:     # also do x < 6
  print(x)
  x = x + 2

leave_loop = False
x = 1

while not leave_loop:    # the not reverses the logic
  print('Loop ran at least once. x = ', x)
  x  = x + 1


  if x == 1:
    print('x == 1')
  if x == 8:
    print('x == 8')
  if x == 10:
     leave_loop = True
```
