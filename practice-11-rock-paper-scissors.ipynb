# Tuples (6, 5.7, 'Hello')

    Tuple is pronounced like "Toople" or like "Tupple"

    Tuple is defined by parentheses ( )

    Tuple is restrictive list

    Tuple is immutable (you can't change elements, order, or size)

    A good reason for being immutable is the need to "unpack" a tuple.


    Using a database: a "tuple" is also known as a "record" ora in a spreadsheet, a "row"

# Assignment 8

```python
"""
This is a 'Rock, Paper, Scissors' game. The game is played between two players(you and computer) where both sides pick rock, paper, or scissors and "show" it at the same time.
                                        Rock wins over scissors, scissors win over paper, and paper wins over rock.
                                                            Are you ready? Let's play the game!

"""

import random

def roll_dice():
    """ Roll three dices and return their values as a tuple """
    die1 = random.randrange(1, 7)
    die2 = random.randrange(1, 7)
    die3 = random.randrange(1, 7)
    return (die1, die2, die3)      # values returned as a tuple

def display_dice(computer_choice):     # computer choice is defined as roll_dice() function in the play_game() function which is the main function that is called
    """ Display the outcome of the dice roll (rock/paper/scissors) """
    x = sum(computer_choice)    # sum function that gets the sum of all three entities and determines if it is rock, paper, or siccors by using if and elif function
    if (x >= 3 and x <= 9):
        return ('r')
    elif (x >= 10 and x <= 15):
        return ('p')
    elif (x >= 16 and x < 19):
        return ('s')

def play_game():
    user_wins = 0
    computer_wins = 0

    print("Welcome to Rock, Paper, and Scissors Game!")
    print("Enter 'r' for rock, 'p' for paper, 's' for scissors, or 'q' to quit.")

    while True:
        user_choice = input("Enter your choice(r/p/s/q): ").lower()

        if user_choice == 'q':
            print("Thank you for playing!")
            break
        elif user_choice == 'r' or user_choice == 'p' or user_choice == 's':
            computer_choice = roll_dice()        # computer choice is defined as roll_dice() function
            computer_result = display_dice(computer_choice)

            print(f"You chose {user_choice}. Computer chose {computer_result}.")

            # Comparison of elements in display-dice() function by the rules of the game
            if (user_choice == 'r' and computer_result == 's') or (user_choice == 'p' and computer_result == 'r') or (user_choice == 's' and computer_result == 'p'):
                print("You win!")
                user_wins += 1                  # running total of user wins
            elif user_choice == computer_result:
                print("It is a tie!")
            else:
                print("Computer wins!")
                computer_wins += 1              # running total of computer wins

            print(f"Score--> You: {user_wins} / Computer: {computer_wins}")
        else:
            print("Invalid value. Please enter r/p/s/q.")

play_game()

```

```python
"""
This is a 'Rock, Paper, Scissors, Lizard, Spock game'. You don't know the rules? 'Oh, it's very simple. Scissors cuts Paper, Paper covers rock, Rock crushes Lizard, Lizard poisons Spock,
   Spock smashes Scissors, Scissors decapitates Lizard, Lizard eats Paper, Paper disproves Spock, Spock vaperizes Rock, and as it always has, Rock crushes Scissors (Sheldon Cooper)'
                                                                         Ready? Let's play!
"""

import random

def roll_dice():
    """ Roll three dices and return their values as a tuple """
    die1 = random.randrange(1, 7)
    die2 = random.randrange(1, 7)
    die3 = random.randrange(1, 7)
    return (die1, die2, die3)

def display_dice(computer_choice):
    """ Display the outcome of the dice roll (rock/paper/scissors/lizard/spock) """
    x = sum(computer_choice)
    if 3 <= x and x <= 5:
        return 'r'
    elif 6 <= x and x <= 8:
        return 'p'
    elif 9 <= x and x <= 12:
        return 's'
    elif 13 <= x and x <= 15:
        return 'l'
    elif 16 <= x and x <= 18:
        return 'k'

def menu():
    print("Welcome to Rock, Paper, Scissors, Lizard, Spock Game!")
    print("Enter 'r' for rock, 'p' for paper, 's' for scissors, 'l' for lizard, 'k' for spock, or 'q' to quit.")

def score(user_choice, computer_result):
    """ Determine the winner of a play based on choices """
    return (
        (user_choice == 'r' and computer_result == 's') or
        (user_choice == 'r' and computer_result == 'l') or
        (user_choice == 'p' and computer_result == 'r') or
        (user_choice == 'p' and computer_result == 'k') or
        (user_choice == 's' and computer_result == 'p') or
        (user_choice == 's' and computer_result == 'l') or
        (user_choice == 'l' and computer_result == 'k') or
        (user_choice == 'l' and computer_result == 'p') or
        (user_choice == 'k' and computer_result == 's') or
        (user_choice == 'k' and computer_result == 'r')
    )

def main():
  menu()  # calling the menu function at the beginning that displays options
  user_wins = 0
  computer_wins = 0

  while True:
      user_choice = input("\nEnter your choice (r/p/s/l/k/q): ").lower()

      if user_choice == 'q':
          print("Thank you for playing!")     # if-elif-else made based on user's input
          break
      elif user_choice == 'r' or user_choice == 'p' or user_choice == 's' or user_choice == 'l' or user_choice == 'k':
          computer_choice = roll_dice()
          computer_result = display_dice(computer_choice)

          print(f"You chose {user_choice}. Computer chose {computer_result}.")

          if score(user_choice, computer_result):   # calling the score() function to determine if user wins and compare it
              print("You win!")
              user_wins += 1
          elif user_choice == computer_result:      # comparing user and computer result to check if it is a tie
              print("It is a tie!")
          else:
              print("Computer wins!")
              computer_wins += 1

          print(f"Score --> You: {user_wins} / Computer: {computer_wins}")   # keeping track of score using f-string
      else:
          print("Try again. Please enter r/p/s/l/k/q.")

if __name__ == '__main__':
    main()
```
