```python
import time

def slow_print(text, delay=0.03):
    for char in text:
        print(char, end='', flush=True)
        time.sleep(delay)
    print()

def intro():
    slow_print("Welcome to the Data Science Maze!")
    slow_print("Three career exits await you: Data Scientist, Data Analyst, and Data Engineer.")
    slow_print("Choose your path wisely...\n")

def first_room():
    slow_print("You are in the Entry Room. Three tunnels appear.")
    slow_print("Question 1: What excites you the most?")
    print("  a) Building predictive models (Forward)")
    print("  b) Analyzing reports and finding insights (Left)")
    print("  c) Creating large data systems (Right)")
    choice = input("Choose (a/b/c): ").lower()
    if choice == 'a':
        return 'forward'
    elif choice == 'b':
        return 'left'
    elif choice == 'c':
        return 'right'
    else:
        slow_print("Invalid choice. Let's try again.\n")
        return first_room()

def second_room(direction):
    slow_print("\nYou move deeper into the maze...")
    slow_print("Another split appears!")
    slow_print("Question 2: Which tools would you be excited to master?")
    if direction == 'forward':
        print("  a) Python and Machine Learning libraries (Forward)")
        print("  b) Business Intelligence tools like Tableau (Left)")
        print("  c) Databases and Hadoop (Right)")
    elif direction == 'left':
        print("  a) Tableau and Excel (Left)")
        print("  b) SQL and Python (Forward)")
        print("  c) Big Data Tools like Spark (Right)")
    elif direction == 'right':
        print("  a) SQL and ETL tools (Right)")
        print("  b) Reporting and Visualization Tools (Left)")
        print("  c) Python for Data Wrangling (Forward)")

    choice = input("Choose (a/b/c): ").lower()
    if choice == 'a':
        return direction  # Continue in same direction
    elif choice == 'b':
        return 'forward'
    elif choice == 'c':
        if direction == 'left':
            return 'right'
        else:
            return 'forward'
    else:
        slow_print("Invalid choice. Let's try again.\n")
        return second_room(direction)

def third_room(direction):
    slow_print("\nAlmost there... One final challenge!")
    slow_print("Question 3: What's your ideal work style?")
    print("  a) Experimenting and innovating new models (Forward)")
    print("  b) Making dashboards and reports for business (Left)")
    print("  c) Building and maintaining pipelines and systems (Right)")

    choice = input("Choose (a/b/c): ").lower()
    if choice == 'a':
        return 'forward'
    elif choice == 'b':
        return 'left'
    elif choice == 'c':
        return 'right'
    else:
        slow_print("Invalid choice. Let's try again.\n")
        return third_room(direction)

def exit_maze(final_direction):
    slow_print("\nYou reach the end of the maze...")
    if final_direction == 'forward':
        slow_print("\U0001F9E0 Congratulations! You found the exit to become a **Data Scientist**!")
    elif final_direction == 'left':
        slow_print("\U0001F4CA Congratulations! You found the exit to become a **Data Analyst**!")
    elif final_direction == 'right':
        slow_print("\U0001F6E0 Congratulations! You found the exit to become a **Data Engineer**!")
    else:
        slow_print("You wander endlessly... Lost in the Maze! (Something went wrong.)")

def play_game():
    intro()
    first_choice = first_room()
    second_choice = second_room(first_choice)
    third_choice = third_room(second_choice)
    exit_maze(third_choice)

if __name__ == "__main__":
    play_game()
```
