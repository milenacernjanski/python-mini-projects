# Slicing

    -- str1[start:stop]
    -- str1[start:stop:step]

```python
str1 = 'Minot State University'
print (str1[0])  # print first character

print(len(str1)) #length of string   22 characters, last character is on 21 place

print(str1[-1])    # print last character of string, second is -2, third is -3
print(str1[21])   # print last character of string

print(str1[0:5])  # print Minot

print(str1[6:11])

print(str1[4:])  # print string starting at 4 until end

print(str1[:4]) # print string starting at 3 to the beginning

print(str1[:]) # print whole string

print(str1[2:-1])  # start at index 2, ends on 21st (without 'y') --- -1 is last character

print(str1[12:-3])  # slice 'Univers'

str1 = 'Minot State University'
print(str1)

print(str1[2:7])

print(str1[2:7:2])

print(str1[6:0:-1])
```

# Assignment 5

```python
while True:
    user_input = input("\nEnter a string or type 'done' to leave the program: ")

    if user_input.lower() == 'done':
        print("\n\nThank you! Bye!")
        break

    print("\nYour string:", user_input)   # Print the user's inputted string to verify it is what the user intended

    reversed_string = user_input[::-1]
    print("This is your string encoded backwards:", reversed_string)  # String backwards

    encoded_index = ''
    for index, char in enumerate(user_input):
        encoded_index += char + str(index)
    print("This is your string with character indices:", encoded_index)  # String with character indices

    encoded_unicode = ''
    for char in user_input:
        encoded_unicode += str(ord(char))
    print("This is your string encoded with unicode numbers:", encoded_unicode) # String with unicode numbers

    encoded_unicode_incremented = ''
    for char in user_input:
        encoded_unicode_incremented += chr(ord(char) + 1)
    print("This is your string with unicode values incremented by 1:", encoded_unicode_incremented)   # String incremented by 1

    string_mess = ''
    for char in user_input:
        for _ in range(5):  # Repeat each character 5 times
            string_mess += char
    print("This is your string turned into a mess:", string_mess)  # Nested loop that turns string into a mess
```
