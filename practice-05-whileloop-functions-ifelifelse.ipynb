## Creating Turtle mi and playing with it as learned in class


```python
import ColabTurtle.Turtle as mi

mi.initializeTurtle()

# set up colors
mi.color('hotpink') # turtle color
mi.bgcolor('slategray') # background color

# draw a square
for i in range(4):
    mi.forward(100)
    mi.right(90)

# move to a new position for the triangle
mi.penup()
mi.forward(150)
mi.right(90)
mi.forward(50)
mi.left(90)
mi.pendown()

# draw a triangle
mi.color('cornflowerblue')
for i in range(3):
    mi.forward(100)
    mi.left(120)

# move to a new position for the rectangle
mi.penup()
mi.backward(200)  # Move left
mi.right(90)      # Face downward
mi.forward(100)   # Move downward
mi.left(90)       # Face left again
mi.pendown()

# draw a rectangle
mi.color('lightsalmon')
length = 200
width = 100
for i in range(2):
    mi.forward(length)
    mi.left(90)
    mi.forward(width)
    mi.left(90)
```

## Using functions, while loop and if-elif-else to create a program


```python
mi.clear()  # to make sure canvas is clear

# define functions for the turtle graphics

def draw_square(x, y, size, color): # draws a square at (x, y) with a given size and color
    mi.penup()
    mi.goto(x, y)
    mi.pendown()
    mi.color(color)
    for i in range(4):
        mi.forward(size)
        mi.right(90)

def draw_rectangle(x, y, width, height, color):  # draws a rectangle at (x, y) with a given width, height, and color
    mi.penup()
    mi.goto(x, y)
    mi.pendown()
    mi.color(color)
    for i in range(2):
        mi.forward(width)
        mi.right(90)
        mi.forward(height)
        mi.right(90)

def draw_triangle(x, y, size, color): # draws a triangle at (x, y) with a given size and color
    mi.penup()
    mi.goto(x, y)
    mi.pendown()
    mi.color(color)
    for i in range(3):
        mi.forward(size)
        mi.right(120)

# menu system - using while and if-elif-else
while True:
    print("\nTurtle Graphics Menu")
    print("1. Draw a square")
    print("2. Draw a rectangle")
    print("3. Draw a triangle")
    print("4. Quit")

    choice = input("Plase select an option (1-4): ")

    if choice == "1":
        print("\nYou chose to draw a square!")
        x = int(input("Enter x-coordinate(200-350): "))   # 200-350 to make sure graphics is visible on the canvas
        y = int(input("Enter y-coordinate(200-350): "))   # 200-350 to make sure graphics is visible on the canvas
        size = int(input("Enter the size of the square: "))
        color = input("Enter the color of the square: ")
        draw_square(x, y, size, color) # calling the function for drawing square

    elif choice == "2":
        print("\nYou chose to draw a rectangle!")
        x = int(input("Enter x-coordinat(200-350): "))
        y = int(input("Enter y-coordinate(200-350): "))
        width = int(input("Enter the width of the rectangle: "))
        height = int(input("Enter the height of the rectangle: "))
        color = input("Enter the color of the rectangle: ")
        draw_rectangle(x, y, width, height, color)  # calling the function for drawing rectangle

    elif choice == "3":
        print("\nYou chose to draw a triangle!")
        x = int(input("Enter x-coordinate(200-350): "))
        y = int(input("Enter y-coordinate(200-350): "))
        size = int(input("Enter the size of the triangle: "))
        color = input("Enter the color of the triangle: ")
        draw_triangle(x, y, size, color)   # calling the function for drawing triangle

    elif choice == "4":
        print("\nGoodbye! Thanks for playing with Turtle Graphics.")
        break

    else:
        print("\nInvalid choice. Please select a valid option.")

# end of program
```
