```python
######################################################################
# Custom Column Transformer
#   -- Handles both numerical and categorical transformations
############################################################ #########

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline

num_attribs = ["longitude", "latitude", "price", "number_of_reviews",
               "number_of_reviews_lmp", "reviews_per_month"]

cat_attribs = ["price"]

num_pipeline = make_pipeline(SimpleImputer(strategy="median"),
                             StandardScaler())

cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore"))

preprocessing = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", cat_pipeline, cat_attribs)])
```

```python
#######################################################################
# If you do not care about naming the trasnformers, you can use:
#       make_column_transformer()
#
#     -- This creates the same code as ColumnTransformer
#         -- Except the transformers are automatically named:
#               pipeline-1     for     "num"
#               pipeline-2     for     "cat"
######################### #############################################

from sklearn.compose import make_column_selector, make_column_transformer

preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)),
    (cat_pipeline, make_column_selector(dtype_include=object)),)
```

```python
######################################################################
# Apply the pipeline transformations to the rentals data
######################################################################

rentals_prepared = preprocessing.fit_transform(rentals)
```

More effective exploration of the hyperparameter space is made possible by randomized search, which selects samples at random from a distribution of hyperparameter values.

I used cross-validation to assess the model's performance after the hyperparameters were adjusted using either grid search or randomized search to make sure the selected hyperparameters would perform well on data that has not yet been observed.
