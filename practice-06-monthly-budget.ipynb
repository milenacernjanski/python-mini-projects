# Extra work on String formatting
    -- Escape characters
        \n       # new line
        \n\n\n   # 3 new lines

        \t  # tab
        \\  # backslash
        \'  # single apostrophe/quote
        \"  # double quote

        

```python
x = 5  # integer
y = 4.3 # float
z = 'My dog Fred'  # string
print (x,y,z)
print (x,y,z, sep='')  # eliminates extra spaces

print(f'I have {x} dogs')   # f-string (modern strings)
print(f'I wish I had {x*3} dogs')

rent= 1000
print(f'rent: ${rent*12} a year')

print(f'rent:\n\n\n${rent*12} a year')   # includes a new line

print(f'rent: \t${rent*12} a year')   # includes a tab

print(f'\n\trent: \t${rent*12} a year')   # includes a \n\t

print('\\')  # print only a backslash   \
print('Fred\'s pet cat')   # print an apostrophe  ' using \'

print('\"This is a quote\"')   # print "character using \"
```

# Special f-string formatting

```python
rent= 1000.2345  # 1000 per month
print(rent)
print(f'rent: ${rent} per month.')

print(f'rent: ${rent:20}')    # sets the column width to 20
print(f'rent: ${rent:25}')    # sets the column width to 25

print(f'rent: ${rent:<25}')    # sets the column width to 25  # < for left justified
print(f'rent: ${rent:>25}')    # sets the column width to 25  # > for right justified

print(f'rent: ${rent:<10.2f}')  # fixed float with two digets after decimal point

print(f'rent: ${rent:<10.2f}{rent*12:<10.2f}')  # shows multiple columns
```

```python
# Monthly budget for Milena Cernjanski

rent = 1500
utilities = 170
gas = 120 # Usually, I fill up my gas tank four times in a month, $30 each time
groceries = 500
gym_membership = 30
other = 200 # Other may include- going out with friends, eating outside, sport activites,...

print('Monthly budget for Milena Cernjanski:')
print('\nrent: $', rent, sep='')
print('utilities: $', utilities, sep='')
print('gas: $', gas, sep='')
print('groceries: $', groceries, sep='')
print('gym_membership: $', gym_membership, sep='')
print('other: $', other, sep='')
print('\n\tTotal: $', rent + utilities + gas + groceries + gym_membership + other, sep='')

answer= input('\nPlease input gas expenses for the month: ')
print (f'Your answer was:', answer)

print('\nMost important expenses to be aware of:')
print(f'\trent: ${rent*12:>5} per year.')
print(f'\tgroceries for year: ${groceries * 12:<10}')
print(f'\tgroceries: ${groceries:<10} groceries per week: {groceries/4}')

print(f'\ngas:\n\t${gas*12:10} a year') # new line + tab + column width changed

print ('\n------------------------')
print('\nWeekly budget:')
print('\tIt is: $', (rent + utilities + gas + groceries + gym_membership + other)/4, sep='')

print('\nYearly budget:')
print(f'\tIt is: $', (rent + utilities + gas + groceries + gym_membership + other)*12, sep='')
```
